```python
import matplotlib.pyplot as plt
import numpy as np

k = np.linspace(.53,1.35,100)
R= .5
B = -4*(k**2-1)*(R-(k**2-1)**2)
fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(6,6))
ax1.plot(k,k*B)
ax1.axhline(y=0, color='r', linestyle='-')
ax1.grid()
plt.savefig("/Users/edwardmcdugald/Research/convection_patterns_wip/figs/writeup/mybb.png")
```
